```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Filling missing values
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unknown")
df['date_added'] = df['date_added'].fillna("Unknown")
df['duration'] = df['duration'].fillna("Unknown")

# Removing duplicates
df.drop_duplicates(inplace=True)
```

Duration Anlaysis

```python
df_movies = df[df['type']=="Movie"].copy()

# Remove rows where duration is "Unknown"
df_movies = df_movies[df_movies['duration'].str.contains("min", na=False)]

# Extract numeric part safely
df_movies['duration_minutes'] = (
    df_movies['duration']
    .str.replace(" min", "", regex=False)
    .astype(int)
)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_movies['duration_minutes'], bins=30, kde=True)
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.savefig("movie_duration.png", bbox_inches='tight', dpi=150)

files.download("movie_duration.png")
plt.show()

```

```python
df_shows = df[df['type']=="TV Show"].copy()
df_shows['seasons'] = df_shows['duration'].str.replace(" Seasons", "").str.replace(" Season", "").astype(int)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df_shows['seasons'])
plt.title("Number of Seasons in TV Shows")
plt.xlabel("Seasons")
plt.ylabel("Count")

plt.savefig("no. of seasons.png", bbox_inches ='tight', dpi = 150)

files.download("no. of seasons.png")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Combine all descriptions into one text
text = " ".join(df['description'].astype(str).tolist())

# remove useless words
stopwords = set(STOPWORDS)
stopwords.update(["Netflix", "series", "movie", "film", "show"])

# Generate the word cloud
wordcloud = WordCloud(
    width=1600,
    height=800,
    background_color="black",
    stopwords=stopwords,
    collocations=False
).generate(text)

# Display the words
plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.savefig("word_cloud.png", bbox_inches ='tight', dpi = 150)

files.download("word_cloud.png")

plt.show()
```

Genre Heatmap


```python
# Split genres
df_genre = df.copy()
df_genre['listed_in'] = df_genre['listed_in'].str.split(", ")

# Explode genres
df_genre = df_genre.explode('listed_in')

genre_pivot = df_genre.pivot_table(
    index='listed_in',
    columns='type',
    aggfunc='size',
    fill_value=0
)
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(genre_pivot, cmap='rocket_r', linewidths=0.5)
plt.title("Genre vs Type Heatmap (Movies vs TV Shows)")
plt.xlabel("Type")
plt.ylabel("Genre")

plt.savefig("genre_heatmap.png", bbox_inches ='tight', dpi = 150)

files.download("genre_heatmap.png")
plt.show()
```
